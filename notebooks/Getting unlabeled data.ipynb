```python
import os
import sys
import pandas as pd
import numpy as np
import glob
import shutil
```

```python
labeled_file = []
with open('../16_12_paddle_detection/16_12_data_detection/Label_16_12.txt') as f:
    for line in f:
        fn = line.split('\t')[0]
        fn = fn.replace('train/', '')
        labeled_file.append(fn)
```

```python
lcd_df = pd.DataFrame(data={'file_name': lcd_imgs, 
                            'prefix': ['../16_12_paddle_detection/22_12/img_lcd'] * len(lcd_imgs)})
```

```python
led7_df = pd.DataFrame(data={'file_name': led7_imgs,
                             'prefix': ['../16_12_paddle_detection/22_12/img_led7'] * len(led7_imgs)})
```

```python
lcd_led7_df = pd.concat([lcd_df, led7_df], axis=0)
```

```python
unlabeled_file = lcd_led7_df[~lcd_led7_df['file_name'].isin(labeled_file)]
```

```python
len(unlabeled_file)
```

```python
unlabeled_file
```

```python
os.makedirs('../img_lcd')
os.makedirs('../img_led7')
```

```python
for r in unlabeled_file.itertuples(index=False):
    fn = r.file_name
    prefix = r.prefix
    
    img_type = prefix.split('/')[-1]
    if img_type.endswith('lcd'):
        shutil.copy(os.path.join(prefix, fn), '../img_lcd')
    else:
        shutil.copy(os.path.join(prefix, fn), '../img_led7')
```
