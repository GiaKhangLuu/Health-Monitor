```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import paddle
```

```python
pretrained_model = os.path.join(ROOT, 'PP-OCR_v3_det_11_29_Teacher', 'best_accuracy')
save_model_dir = './text_detection_model_teacher_dml_151222/'
num_epoch = 10
eval_batch_step = 200  # evaluation is run every 200 iterations after the 200th iteration
train_data_dir = os.path.join(ROOT, 'data')
label_file_list = os.path.join(ROOT, 'data', 'screen_data_051222', 'Label.txt')
use_gpu=False
teacher1_params = os.path.join(ROOT, 'teacher1.pdparams')
teacher2_params = os.path.join(ROOT, 'teacher2.pdparams')
ratio_list=1.0
```

```python
# load pretrained model
all_params = paddle.load('{}.pdparams'.format(pretrained_model))
#paddle.save(s_params, "./teacher.pdparams")
s1_params = {key[len("Student."):]: all_params[key] for key in all_params if "Student." in key}
paddle.save(s1_params, teacher1_params)
s2_params = {key[len("Student2."):]: all_params[key] for key in all_params if "Student2." in key}
paddle.save(s2_params, teacher2_params)
```
